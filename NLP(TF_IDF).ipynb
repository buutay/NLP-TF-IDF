## TF-векторы для множества документов

```python
docs =["The faster Harry got to the store, the faster and faster Harry would get home."]
docs.append("Harry is hairy and faster than Jill.")
docs.append("Jill is not as hairy as Harry.")
```

```python
doc_tokens = []
for doc in docs:
  doc_tokens += [sorted(tokenizer.tokenize(doc.lower()))]
all_doc_tokens = sum(doc_tokens, [])
lexicon = sorted(set(all_doc_tokens))
```

```python
len(doc_tokens[0])
```

```python
len(all_doc_tokens)
```

```python
len(lexicon)
```

```python
lexicon
```

```python
from collections import OrderedDict
zero_vector = OrderedDict((token, 0) for token in lexicon)
zero_vector
```

```python
import copy
doc_vectors = []
for doc in docs:
  vec = copy.copy(zero_vector)
  tokens = tokenizer.tokenize(doc.lower())
  token_counts = Counter(tokens)
  for key, value in token_counts.items():
    vec[key] = value/len(lexicon)
  doc_vectors.append(vec)
doc_vectors
```

##Коэффициент Отиаи

```python
import math
def cosine_sim(vec1, vec2):
  vec1 = [val for val in vec1.values()]
  vec2 = [val for val in vec2.values()]
  dot_prod = 0
  for i, v in enumerate(vec1):
    dot_prod += v * vec2[i]

  mag_1 = math.sqrt(sum([x**2 for x in vec1]))
  mag_2 = math.sqrt(sum([x**2 for x in vec2]))
  return dot_prod / (mag_1 * mag_2)
```

```python
cosine_sim(doc_vectors[0], doc_vectors[1])
```

```python
cosine_sim(doc_vectors[0], doc_vectors[2])
```

```python
cosine_sim(doc_vectors[1], doc_vectors[2])
```

#3. Закон Ципфа

Закон Ципфа гласит, что в некотором корпусе высказываний на естественном языке частотность любого слова обратно пропорциональна его позиции в таблице частотностей.

Таким образом, при достаточно большом корпусе можно на основе этого разбиения предсказать статистическую вероятность вхождения данного слова в любом из документов корпуса.

```python
nltk.download('brown')
from nltk.corpus import brown
brown.words()[:10]
```

```python
len(brown.words())
```

##Ранжирование по релевантности

```python
document_tfidf_vectors = []
for doc in docs:
  vec = copy.copy(zero_vector)
  tokens = tokenizer.tokenize(doc.lower())
  token_counts = Counter(tokens)

  for key, value in token_counts.items():
    docs_containing_key = 0
    for doc in docs:
      if key in doc:
        docs_containing_key += 1
      tf = value / len(lexicon)
      if docs_containing_key:
        idf = len(docs) / docs_containing_key
      else:
        idf = 0
      vec[key] = tf * idf
  document_tfidf_vectors.append(vec)
```

```python
query = "How long does it take to get to the store?"
query_vec = copy.copy(zero_vector)
tokens = tokenizer.tokenize(query.lower())
token_counts = Counter(tokens)
for key, value in token_counts.items():
  docs_containing_key = 0
  for doc in docs:
    if key in doc.lower():
      docs_containing_key += 1
  if docs_containing_key == 0:
    continue
    tf = value / len(tokens)
    idf = len(docs) / docs_containing_key
  query_vec[key] = tf * idf
```
